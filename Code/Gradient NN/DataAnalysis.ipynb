```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("test.txt", sep=" ")
```

```python
# Features (inputs)
X = df.drop(columns=["Radius"])

# Target (output)
y = df["Radius"]

# Print shape of data
print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
```

```python
sns.pairplot(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation with Radius")
plt.show()
```

```python
for col in X.columns:
    plt.figure(figsize=(5, 4))
    sns.regplot(x=df[col], y=df["Radius"])
    plt.title(f"Correlation between {col} and Radius")
    plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train a basic model
model = RandomForestRegressor()
model.fit(X, y)

# Plot feature importance
importances = model.feature_importances_
plt.figure(figsize=(6, 4))
sns.barplot(x=importances, y=X.columns)
plt.title("Feature Importance for Predicting Radius")
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_regression

mi_scores = mutual_info_regression(X, y)
mi_scores = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)

# Plot results
plt.figure(figsize=(6, 4))
sns.barplot(x=mi_scores.values, y=mi_scores.index)
plt.title("Mutual Information Scores for Radius")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation = df.corr()["Radius"].drop("Radius")
print("Feature correlations with Radius:\n", correlation)

# Visualize correlation
plt.figure(figsize=(5, 4))
sns.barplot(x=correlation.values, y=correlation.index)
plt.title("Feature Correlation with Radius")
plt.show()
```
